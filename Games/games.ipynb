# Rock Paper Scissors

```python
import random
def playgame(p1, p2):
    if p1 == p2:
        return 2 #TIE
    if p1 == 'Rock':
        if p2 == 'Paper':
            return 1 #p2 win
        else:
            return 0 #p1 win
    elif p1 == 'Paper':
        if p2 == 'Scissors':
            return 1
        else:
            return 0
    else:
        if p2 == 'Rock':
            return 1
        else:
            return 0

def play():
    while True:
        player = input('Rock, Paper, or Scissors?')
        cpu = random.randint(0,2)
        cpu_choice = ''
        if cpu == 0:
            cpu_choice = 'Rock' 
        elif cpu == 1:
            cpu_choice = 'Paper'
        else:
            cpu_choice = 'Scissors' 
        result = playgame(player, cpu_choice)
        if result == 2:
            print(f"It's a Tie! You chose {player}, the cpu chose {cpu_choice}\n")
        if result == 1:
            print(f"The CPU Won! You chose {player}, the cpu chose {cpu_choice}\n")
        if result == 0:
            print(f"You Won! You chose {player}, the cpu chose {cpu_choice}\n")
        playagain = input('Would you like to play again? (y/n)')
        if playagain == 'n':
            break

play()
        

    
```

# Dots and Boxes

```python
from tkinter import *
import numpy as np

class DotsandBoxes:
    def __init__(self, rows=3, cols=3):
        self.window = Tk()
        self.window.title("Dots and Boxes")
        self.rows = rows
        self.cols = cols
        self.current_player = 1
        self.scores = {1: 0, 2: 0}
        self.create_board()
        self.create_gui()
    def play(self):
        self.window.mainloop()
    
    def create_board(self):
        self.board = np.zeros((2 * self.rows + 1, 2 * self.cols + 1), dtype=int)
    
    def create_gui(self):
        self.canvas = Canvas(self.window, width = 400, height = 400)
        self.canvas.pack()
        self.canvas.bind("<Button-1>", self.on_canvas_click)
        self.draw_grid()
    def draw_grid(self):
        for i in range(1, 2*self.rows+1, 2):
            for j in range(1, 2*self.cols +1, 2):
                self.canvas.create_oval(j * 40 - 5, i * 40 - 5, j * 40 + 5, i * 40 + 5, fill="black")
    def on_canvas_click(self, event):
        col = event.x // 40
        row = event.y // 40

        if self.board[row][col] == 0:
            self.board[row][col] = self.current_player
            self.check_for_box(row, col)
            self.switch_player()
            self.draw_move(row, col)
    def check_for_box(self, row, col):
        if row % 2 == 1 and col % 2== 1:
            if self.board[row-1][col] != 0 and self.board[row+1][col] != 0 and self.board[row][col-1] != 0 and self.board[row][col+1] != 0:
                self.scores[self.current_player] += 1
    def switch_player(self):
        self.current_player = 3 - self.current_player
    def draw_move(self, row, col):
        if row % 2 == 0:
            self.canvas.create_line(col*40, row*40 + 20, (col + 2) * 40, row * 40 + 20, width = 2, fill = "blue")
        else:
            self.canvas.create_line(col*40 + 20, row*40, col * 40 + 20, (row + 2) * 40, width = 2, fill="red")
        self.check_game_over()
    def check_game_over(self):
        if np.all(self.board != 0):
            winner = max(self.scores, key=self.scores.get)
            print(f'Game Over! Player {winner} wins with a total score of {self.scores[winner]}')
    
game1 = DotsandBoxes()
game1.play()
```
